```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
LR.fit(df1[['YearsExperience']],df1['Salary'])
```

```python
Y_pred = LR.predict(df1[['YearsExperience']])
```

```python
from sklearn.metrics import r2_score
R = r2_score(Y_pred,df1['Salary'])
R
```

```python
import matplotlib.pyplot as plt
plt.scatter(df1['YearsExperience'],df1['Salary'])
plt.plot(df1['YearsExperience'],Y_pred)
plt.xlabel("'YearsExperience'")
plt.ylabel('Salary')
```
